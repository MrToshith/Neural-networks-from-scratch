# Neural Network Implementation from Scratch (MNIST)

### Part-1: Data Preparation

#### Step-1: Importing Libraries and Loading Data

```python
import numpy as np
import pandas as pd

data = pd.read_csv('data.csv')
data.head()
```

#### Step-2: Data Conversion

```python
data = np.array(data)
m,n = data.shape
```

#### Step-3: Splitting and Normalization

```python
train_data = data[0:int(0.8*m),:]
test_data = data[int(0.8*m):,:]
```

```python
X_train = train_data[:,1:].T
X_train = X_train / 255
y_train = train_data[:,0]

X_test = test_data[:,1:].T
X_test = X_test / 255
y_test = test_data[:,0]
```

### Part-2: Neural Network Architecture

#### Step-4: Parameter Initialization

```python
def initialilization(input_size=784, hidden_layer_size=10, output_layer_size=10):
    W1 = np.random.rand(hidden_layer_size, input_size) * 0.01
    B1 = np.random.rand(hidden_layer_size,1) * 0.01
    W2 = np.random.rand(output_layer_size, hidden_layer_size) * 0.01
    B2 = np.random.rand(output_layer_size,1) * 0.01
    return W1,B1,W2,B2
```

#### Step-5: Activation Functions

```python
def ReLU(a):
    return np.maximum(0,a)

def Softmax(a):
    return np.exp(a) / sum(np.exp(a))
```

#### Step-6: Forward Propagation

```python
def forward_propagation(W1,B1,W2,B2,X):
    Z1 = W1.dot(X) + B1
    A1 = ReLU(Z1)
    Z2 = W2.dot(A1) + B2
    A2 = Softmax(Z2)
    return Z1,A1,Z2,A2
```

### Part-3: Optimization Algorithms

#### Step-7: Helper Functions

```python
def one_hot_encoding(y):
    m = y.size
    output_y = np.zeros((10,m))
    output_y[y,np.arange(m)] = 1
    return output_y
```

#### Step-8: Backward Propagation

```python
def backward_propagation(W2,Z1,A1,A2,X,Y):
    dZ2 = A2 - one_hot_encoding(Y)
    dW2 = 1 / m * (dZ2.dot(A1.T))
    dB2 = 1 / m * (np.sum(dZ2))
    dZ1 = (W2.T).dot(dZ2) * (Z1 > 0)
    dW1 = 1 / m * (dZ1.dot(X.T))
    dB1 = 1 / m * (np.sum(dZ1))
    return dW2,dB2,dW1,dB1
```

#### Step-9: Parameter Updates

```python
def param_update(dW2,dB2,dW1,dB1,alpha,W1,B1,W2,B2):
    W1 = W1 - alpha*dW1
    B1 = B1 - alpha*dB1
    W2 = W2 - alpha*dW2
    B2 = B2 - alpha*dB2
    return W1,B1,W2,B2
```

### Part-4: Training and Evaluation

#### Step-10: Training Loop

```python
def train_model(iterations,alpha,X,y):
    W1,B1,W2,B2 = initialilization()
    for i in range(iterations):
        Z1,A1,Z2,A2 = forward_propagation(W1,B1,W2,B2,X)
        dW2,dB2,dW1,dB1 = backward_propagation(W2,Z1,A1,A2,X,y)
        W1,B1,W2,B2 = param_update(dW2,dB2,dW1,dB1,alpha,W1,B1,W2,B2)
    return W1,B1,W2,B2
```

#### Step-11: Testing

```python
def predict(W1,B1,W2,B2,X):
    Z1,A1,Z2,A2 = forward_propagation(W1,B1,W2,B2,X)
    return np.argmax(A2, axis=0)

def accuracy(predictions,y):
    return np.mean(predictions==y) * 100
```

```python
W1,B1,W2,B2 = train_model(1000,0.1,X_train,y_train)
predictions = predict(W1,B1,W2,B2,X_test)
test_accuracy = accuracy(predictions,y_test)
print("Accuracy : ", test_accuracy)
```
